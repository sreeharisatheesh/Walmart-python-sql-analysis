```python
import pandas as pd
df = pd.read_csv("Walmart.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.duplicated().sum()
df.drop_duplicates()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.unit_price  = df.unit_price.apply(lambda x:x[1:]).astype(float)
```

```python
df['total'] = df['quantity']*df['unit_price']
```

```python
df
```

```python
df.to_csv('walmart_clean.csv',index=False)
```

```python

engine_mysql = create_engine('mysql+pymysql://root:root@localhost:3306/walmart')
try:
    engine_mysql
    print('mysql connection successful')
except:
    print('unable to connect')
```
